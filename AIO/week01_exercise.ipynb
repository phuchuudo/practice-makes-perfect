```python
def checktype_int(value):
  """
  Kiểm tra kiểu dữ liệu của giá trị nhập vào có phải là kiểu int hay không.

  Args:
      value: Giá trị cần kiểm tra.

  Returns:
      True nếu kiểu dữ liệu của giá trị là int, False nếu không.
  """
  return type(value) == int

def ex1(tp = 2, fp = 3, fn = 4):
  """
  Tính toán độ chính xác (precision), độ nhạy (recall) và F1-score dựa trên số lượng TP, FP và FN.

  Args:
      tp (int, optional): Số lượng True Positive. Defaults to 2.
      fp (int, optional): Số lượng False Positive. Defaults to 3.
      fn (int, optional): Số lượng False Negative. Defaults to 4.
  """
  ok = True
  if checktype_int(tp) == False:
    print('tp must be int')
    ok = False
  if checktype_int(fp) == False:
    print('fp must be int')
    ok = False
  if checktype_int(fn) == False:
    print('fn must be int')
    ok = False
  if ok == False:
    return

  if tp <= 0 or fp <= 0 or fn <= 0:
    print('tp and fp and fn must be greated than zero')
    return
  precision = tp / (tp + fp)
  recall = tp / (tp + fn)
  f1_score = 2 * (precision * recall) / (precision + recall)
  print(f'precision is {round(precision, 3)}, recall is {round(recall, 3)} and f1_score is {round(f1_score, 3)}')

ex1()
ex1(3,0, 5)
```

## TP, FP, FN trong việc đánh giá mô hình phân loại

TP, FP và FN là những thuật ngữ phổ biến được sử dụng để đánh giá hiệu suất của mô hình phân loại. Chúng đại diện cho các loại lỗi và dự đoán chính xác được thực hiện bởi mô hình.

**1. TP (True Positive):** Dự đoán chính xác là dương tính (positive). Mô hình dự đoán đúng một mẫu thuộc lớp dương tính.

**2. FP (False Positive):** Dự đoán sai là dương tính (positive). Mô hình dự đoán một mẫu thuộc lớp dương tính, nhưng thực tế nó thuộc lớp âm tính.

**3. FN (False Negative):** Dự đoán sai là âm tính (negative). Mô hình dự đoán một mẫu thuộc lớp âm tính, nhưng thực tế nó thuộc lớp dương tính.

**Bảng tóm tắt:**

|  | Dự đoán là dương tính | Dự đoán là âm tính |
|---|---|---|
| Thực tế là dương tính | **TP (True Positive)** | **FN (False Negative)** |
| Thực tế là âm tính | **FP (False Positive)** | **TN (True Negative)** |

**Ví dụ:**

* **Mô hình phân loại spam:**
    * TP: Mô hình dự đoán đúng một email là spam.
    * FP: Mô hình dự đoán sai một email là spam, trong khi thực tế nó không phải là spam.
    * FN: Mô hình dự đoán sai một email không phải là spam, trong khi thực tế nó là spam.

**Lưu ý:**

* **TN (True Negative):**  Dự đoán chính xác là âm tính (negative). Mô hình dự đoán đúng một mẫu thuộc lớp âm tính.
* Các thuật ngữ TP, FP, FN và TN được sử dụng để tính toán các chỉ số đánh giá mô hình phân loại như độ chính xác (accuracy), độ nhạy (sensitivity), độ đặc hiệu (specificity), F1-score...
* Việc lựa chọn chỉ số đánh giá phù hợp phụ thuộc vào mục tiêu của bài toán phân loại.

**Kết luận:**

Hiểu rõ các thuật ngữ TP, FP và FN là rất quan trọng để đánh giá hiệu suất của mô hình phân loại và lựa chọn mô hình phù hợp với nhu cầu của bài toán.


```python
import math

# check is number
def is_number(n):
  """
  This function checks if a given input is a number.

  Args:
    n: The input to check.

  Returns:
    True if the input is a number, False otherwise.
  """
  try:
    float(n)
  except ValueError:
    return False
  return True

def sig(x = 1.5):
  """
  This function calculates the sigmoid activation function.

  Args:
    x: The input value.

  Returns:
    The sigmoid value.
  """
  return 1 / (1 + math.e ** (-x))

def relu(x = 1.5):
  """
  This function calculates the ReLU activation function.

  Args:
    x: The input value.

  Returns:
    The ReLU value.
  """
  if x <= 0:
    return 0
  return x

def elu(x = 1.5, alpha = 0.01):
  """
  This function calculates the ELU activation function.

  Args:
    x: The input value.
    alpha: The alpha value for the ELU function.

  Returns:
    The ELU value.
  """
  if x <= 0:
    return alpha * (math.e ** x - 1)
  return x

def ex2():
  """
  This function takes user input for an x value and activation function,
  then calculates and prints the corresponding activation value.
  """
  inp = input("input x = ")
  if is_number(inp) == False:
    print('x must be anumber')
    return

  x = float(inp)

  name_function = input("Input activation Function (sigmoid|relu|elu): ")
  if name_function != "sigmoid" and name_function != "relu" and name_function != "elu":
    print(f'{name_function} is not supported')
    return

  if name_function == "sigmoid":
    print(f'sigmoid: f({x}) = {sig(x)}')
    return
  if name_function == "relu":
    print(f'relu: f({x}) = {relu(x)}')
    return
  if name_function == "elu":
    print(f'elu: f({x}) = {elu(x)}')
    return

ex2()
```

## So sánh ưu nhược điểm của các hàm kích hoạt Sigmoid, ReLU và ELU

### Sigmoid

**Ưu điểm:**

* **Dễ hiểu và dễ triển khai:** Hàm Sigmoid là một hàm đơn giản, dễ hiểu và dễ triển khai.
* **Kết quả đầu ra nằm trong khoảng [0, 1]:** Điều này giúp dễ dàng diễn giải kết quả đầu ra của mô hình, ví dụ như xác suất của một sự kiện.

**Nhược điểm:**

* **Vấn đề gradient vanishing:** Khi giá trị đầu vào lớn hoặc nhỏ, gradient của hàm Sigmoid sẽ rất nhỏ, dẫn đến việc gradient vanishing và khó khăn trong việc huấn luyện mô hình.
* **Kết quả đầu ra không đối xứng:**  Điều này có thể gây ra vấn đề trong việc huấn luyện mô hình, đặc biệt là khi dữ liệu không cân bằng.
* **Tính toán chậm:** So với ReLU, Sigmoid cần nhiều thời gian tính toán hơn.

### ReLU (Rectified Linear Unit)

**Ưu điểm:**

* **Giải quyết vấn đề gradient vanishing:** ReLU có gradient là 1 khi đầu vào dương và 0 khi đầu vào âm, giúp tránh vấn đề gradient vanishing.
* **Tính toán nhanh:** ReLU đơn giản và nhanh chóng hơn Sigmoid.
* **Hiệu suất tốt hơn:** Trong nhiều trường hợp, ReLU cho kết quả tốt hơn Sigmoid.

**Nhược điểm:**

* **Vấn đề "dead neuron":** Khi đầu vào âm, ReLU sẽ trả về 0 và gradient của nó cũng là 0. Điều này có thể dẫn đến "dead neuron" (nơ-ron chết) và không đóng góp gì vào việc học.
* **Không có giá trị âm:**  Điều này có thể gây khó khăn trong việc xử lý dữ liệu có giá trị âm.

### ELU (Exponential Linear Unit)

**Ưu điểm:**

* **Giải quyết vấn đề "dead neuron":** ELU có giá trị âm khi đầu vào âm, giúp tránh vấn đề "dead neuron".
* **Gradient luôn dương:** Gradient của ELU luôn dương, giúp cải thiện tốc độ học.
* **Kết quả đầu ra đối xứng:** ELU cho kết quả đầu ra đối xứng, giúp xử lý dữ liệu không cân bằng tốt hơn.

**Nhược điểm:**

* **Tính toán phức tạp hơn:** So với ReLU, ELU phức tạp hơn và cần nhiều thời gian tính toán hơn.

**Kết luận:**

* **ReLU** là lựa chọn phổ biến cho các lớp ẩn của mạng nơ-ron do hiệu suất cao và tốc độ tính toán nhanh.
* **ELU** là một lựa chọn tốt hơn ReLU khi gặp vấn đề "dead neuron" và xử lý dữ liệu không cân bằng.
* **Sigmoid** thường được sử dụng trong lớp đầu ra của mô hình phân loại để tạo ra xác suất.

**Lưu ý:** Lựa chọn hàm kích hoạt phù hợp phụ thuộc vào cấu trúc của mô hình, dữ liệu đầu vào và mục tiêu của bài toán.


```python
import random
import math

def RMSE(n, preds, targets):
  """
  This function calculates the Root Mean Squared Error (RMSE) between predictions and targets.

  Args:
    n: The number of samples.
    preds: A list of predictions.
    targets: A list of target values.

  Returns:
    The RMSE value.
  """
  sum = 0
  for i in range(n):
    sum += (preds[i] - targets[i]) ** 2
  return math.sqrt(sum / n)

def MSE(n, preds, targets):
  """
  This function calculates the Mean Squared Error (MSE) between predictions and targets.

  Args:
    n: The number of samples.
    preds: A list of predictions.
    targets: A list of target values.

  Returns:
    The MSE value.
  """
  sum = 0
  for i in range(n):
    sum += (preds[i] - targets[i]) ** 2
  return sum / n

def MAE(n, preds, targets):
  """
  This function calculates the Mean Absolute Error (MAE) between predictions and targets.

  Args:
    n: The number of samples.
    preds: A list of predictions.
    targets: A list of target values.

  Returns:
    The MAE value.
  """
  sum = 0
  for i in range(n):
    sum += abs(preds[i] - targets[i])
  return sum / n

def is_number_int(n):
  """
  This function checks if a given input is an integer.

  Args:
    n: The input to check.

  Returns:
    True if the input is an integer, False otherwise.
  """
  try:
    int(n)
  except ValueError:
    return False
  return True

def ex3():
  """
  This function generates random predictions and targets, calculates different loss metrics,
  and prints the results.
  """
  inp = input('Input number of samples (integer number) which are generated: ')
  if is_number_int(inp) == False:
    print('number of samples must be an integer number')
    return
  n = int(inp)
  loss_name = input('Input loss name: ')
  preds = []
  targets = []
  for i in range(n):
    pred = random.uniform(0, 10)
    target = random.uniform(0, 10)
    loss_i = (pred - target) ** 2
    if loss_name == 'MAE':
      loss_i = abs(pred - target)
    print(f'loss name: {loss_name}, sample: {i}, pred: {pred}, target: {target}, loss: {loss_i}')
    preds.append(pred)
    targets.append(target)

  loss_final = 0
  if loss_name == 'RMSE':
    loss_final = RMSE(n, preds, targets)

  if loss_name == 'MSE':
    loss_final = MSE(n, preds, targets)

  if loss_name == 'MAE':
    loss_final = MAE(n, preds, targets)
  print(f'final {loss_name}: {loss_final}')

ex3()
```

## So sánh Mean Absolute Error (MAE) và Mean Difference of nth Root Error

Mặc dù thuật ngữ "Mean Difference of nth Root Error" (Sai số Trung bình của Hiệu các Căn bậc n) không phải là một thước đo tiêu chuẩn trong học máy, chúng ta có thể hiểu ý nghĩa của nó và so sánh nó với MAE dựa trên mô tả của nó. Hãy giả sử "Mean Difference of nth Root Error" đề cập đến trung bình của các hiệu số giữa căn bậc n của các giá trị dự đoán và giá trị thực tế.

**Mean Absolute Error (MAE)**

* **Công thức:** MAE = (1/n) * Σ |y_i - ŷ_i|
* **Diễn giải:** MAE đo lường trung bình của sai số tuyệt đối giữa các giá trị dự đoán (ŷ_i) và giá trị thực tế (y_i). Nó biểu thị trung bình độ lớn của các sai số, bất kể hướng của chúng (quá ước lượng hoặc đánh giá thấp).

**Mean Difference of nth Root Error (giả định)**

* **Công thức (giả sử căn bậc n của cả giá trị dự đoán và thực tế):**  (1/n) * Σ |y_i^(1/n) - ŷ_i^(1/n)|
* **Diễn giải:** Thước đo này đo lường trung bình của hiệu số giữa căn bậc n của các giá trị dự đoán và giá trị thực tế. Về cơ bản, nó đo lường độ lệch của căn bậc n của các dự đoán so với căn bậc n của các giá trị thực tế.

**So sánh:**

| Thước đo | Ưu điểm | Nhược điểm |
|---|---|---|
| MAE | Dễ hiểu và giải thích.  Khắc phục được outlier. | Nhạy cảm với thang đo của dữ liệu. |
| Mean Difference of nth Root Error | Cung cấp thước đo sai số ở căn bậc n của giá trị.  Có thể nhạy cảm hơn với lỗi nhỏ. |  Ít phổ biến và ít được hiểu rõ hơn.  Có thể khó diễn giải. |

**Sự khác biệt chính:**

* **Thang đo:**  MAE nhạy cảm với thang đo của dữ liệu, trong khi thước đo sai số căn bậc n có thể ít nhạy cảm hơn, đặc biệt là đối với giá trị lớn. Điều này là do việc khai căn bậc n của một số lớn có xu hướng kéo độ lớn của nó gần hơn với các giá trị khác.
* **Độ nhạy:** MAE được khắc phục outlier, vì nó xem xét giá trị tuyệt đối của sai số.  Sai số căn bậc n có thể nhạy cảm hơn với outlier, vì sự khác biệt lớn trong căn bậc n có thể bị khuếch đại bởi phép toán căn bậc n.
* **Diễn giải:** MAE dễ dàng diễn giải vì nó đại diện cho trung bình độ lớn của sai số.  Sai số căn bậc n có thể khó diễn giải hơn vì nó đo lường sự khác biệt giữa căn bậc n của các giá trị, điều này có thể không có ý nghĩa trực quan rõ ràng.

**Khi sử dụng thước đo nào:**

* **MAE:** Sử dụng MAE khi bạn cần một thước đo lỗi đơn giản và vững chắc, đặc biệt là nếu dữ liệu của bạn chứa outlier.
* **Mean Difference of nth Root Error:** Sử dụng thước đo này (nếu phù hợp với vấn đề của bạn) khi bạn quan tâm đến sai số ở căn bậc n của các dự đoán của bạn. Điều này có thể hữu ích khi xử lý các biến có độ lệch cao hoặc tuân theo phân phối lũy thừa.

**Lưu ý:**  "Mean Difference of nth Root Error" là một thước đo giả định. Điều quan trọng là phải xem xét cẩn thận thước đo nào phù hợp nhất với nhu cầu và dữ liệu cụ thể của bạn.




```python
# trac nghiem 1
import math
def calc_f1_score(tp, fp, fn):
  precision = tp / (tp + fp)
  recall = tp / (tp + fn)
  f1_score = 2 * (precision * recall) / (precision + recall)
  return f1_score

assert round(calc_f1_score(tp=2, fp=3, fn=5), 2) == 0.33
print(round(calc_f1_score(tp=2, fp=4, fn=5), 2))
```
