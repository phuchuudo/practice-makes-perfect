```python
# Define lists for Can and Chi elements
can = ["Canh", "Tân", "Nhâm", "Quý", "Giáp", "Ất", "Bính", "Đinh", "Mậu", "Kỷ"]
chi = ["Thân", "Dậu", "Tuất", "Hợi", "Tý", "Sửu", "Dần", "Mão", "Thìn", "Tỵ", "Ngọ", "Mùi"]

# Define a function to calculate the Can Chi calendar year
def calculate_can_chi_calendar(year):
    """
    Calculates the Can Chi calendar year for a given year.

    Args:
        year: The year to calculate the Can Chi calendar year for.

    Returns:
        A string representing the Can Chi calendar year.
    """
    result = ''
    # Calculate the Can element based on the year modulo 10
    result += can[year % 10]
    # Add a space and the Chi element based on the year modulo 12
    result += ' ' + chi[year % 12]
    return result

# Print the Can Chi calendar year for the given years
print(calculate_can_chi_calendar(2024))
print(calculate_can_chi_calendar(1990))
print(calculate_can_chi_calendar(2023))
print(calculate_can_chi_calendar(1997))
```

```python
import nltk
from nltk.tokenize import word_tokenize

# Download necessary resources for tokenization and POS tagging
nltk.download('punkt')  # Download the Punkt sentence tokenizer
nltk.download('averaged_perceptron_tagger')  # Download the Averaged Perceptron tagger

# Sample text data
data = "Tôi thích học AI và toán."

# Step 1: Tokenization
# Use word_tokenize() to split the text into individual words
tokenization = word_tokenize(data)
# This will result in: ['Tôi', 'thích', 'học', 'AI', 'và', 'toán', '.']

# Step 2: Part-of-Speech (POS) Tagging
# Use nltk.pos_tag() to assign grammatical tags (e.g., noun, verb, adjective) to each word
result = nltk.pos_tag(tokenization)
# This will output a list of tuples, where each tuple is a word and its POS tag:
# [('Tôi', 'PRP'), ('thích', 'VB'), ('học', 'NN'), ('AI', 'NNP'), ('và', 'CC'), ('toán', 'NN'), ('.', '.')]

print(result)
```
