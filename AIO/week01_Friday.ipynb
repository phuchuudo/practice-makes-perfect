```python
# pi estimation
# Gregory-Leibniz series
n = 1000000
PI = 0
for i in range(1, n):
  PI += (-1)**(i+1) / (2 * i - 1)
PI = 4 * PI
print(f'PI estimation: {PI}')

#Nilakantha series
n = 1000000
PI = 0
for i in range(0, n):
  PI += (-1)**(i) / ((2*i+2)*(2*i+3)*(2*i+4))
PI = 4 * PI + 3
print(f'PI estimation: {PI}')
```

```python
# e estimation
n = 100000
e_val = (1 + 1/n) ** n
print(e_val)
```

```python
# sqrt estimation
def sqrt_estimation(N, num_loops = 100):
  x_n = N / 2.0
  for i in range(num_loops):
    x_np1 = (x_n + N / x_n) / 2.0
    x_n = x_np1
  return x_np1

print(sqrt_estimation(9))
print(sqrt_estimation(2))
```
