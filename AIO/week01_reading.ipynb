```python
# F-string
student_name = "Tom"
class_name = "AI VietNam"
message = f"{student_name} học lập trình tại {class_name}"
print(message)
n = len(message)
print(f"Độ dài chuỗi là {n}")
print(f"Phần từ đầu tiên là {message[0]} và phân tử cuối cùng là {message[-1]}")
```

```python
# list comprehansion
squares = [x ** 2 for x in range(10) if x % 2 == 0]
print(squares)
```

```python
# dictionary comprehansion
squares_dict = {x : x ** 2 for x in range(10)}
print(squares_dict)
```

```python
# for and continue
for i in range(5):
  if i == 2:
    continue
  if i == 4:
    break
  print(i)
```

```python
# for lồng
matrix = [
    [1, 2, 3],
    [3, 4, 5],
    [5, 6, 8]
]

for i in range(len(matrix)):
  for j in range(len(matrix[i])):
    print(f'value vị trí {i} và {j} là : {matrix[i][j]}')
```

```python
#while
i = 1
while i <= 10:
  i += 1
  if i % 2 == 0:
    continue
  if i > 7:
    break
  print(i)
```
